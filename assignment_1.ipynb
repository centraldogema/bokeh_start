## Reading and displaying data

```python
import pandas as pd
import numpy as np
df = pd.read_csv("oil_consumption_data.csv", index_col = 0, na_values = "-")
colsAsInt = map(int, list(df))

df
```

## Intializing the Jupyter notebook

```python
from bokeh.io import output_notebook
output_notebook()
```

## (Messy) Representation of Every Country with Interactive MouseOver

Graph of all data contained within the dataset with interactive hover tool.

```python
from bokeh.models import HoverTool
from bokeh.plotting import figure, show

hover = HoverTool(tooltips=[
    ("index", "$index"),
    ("(x,y)", "($x, $y)")
])

p = figure(plot_width = 800, plot_height = 400, tools = [hover], title = "With Hover")
p.left[0].formatter.use_scientific = False
for country in df.index:
    y = df.loc[country]
    p.circle(colsAsInt, y, fill_color = "white")
    p.line(colsAsInt, y)
    
show(p, notebook_handle = True)
```

## Generating an interactive line graph of the first 11 countries

More or less simulating the interactive legends page from http://bokeh.pydata.org/en/latest/docs/user_guide/interaction/legends.html#userguide-interaction-legends.

```python
from bokeh.palettes import Spectral11

oil_data_plot = figure(plot_width=800, plot_height=400, title = "Click on the legend to hide countries.")
oil_data_plot.left[0].formatter.use_scientific = False

def updateGraph(countries):
    for country, color in zip(countries, Spectral11):
        y = df.loc[country]
        oil_data_plot.circle(colsAsInt, y, color = color, fill_color = "white", legend = country)
        oil_data_plot.line(colsAsInt, y, color = color, legend = country)
    oil_data_plot.x_range.end = 2025
    oil_data_plot.legend.location = "top_right"
    oil_data_plot.legend.click_policy="hide"

def getRandomCountries():
    randArr = []
    while len(randArr) < 11:
        random = np.random.randint(0, len(df.index))
        if random not in randArr:
            randArr.append(random)
    randomCountries = []
    for i in randArr:
        randomCountries.append(df.index[i])
    return randomCountries
    
updateGraph(getRandomCountries())
show(oil_data_plot, notebook_handle = True)
```
