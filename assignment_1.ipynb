## Analyzing total oil consumption by country.

Source: http://docs.google.com/spreadsheet/pub?key=pyj6tScZqmEcm0fIa0IVtKw&output=xlsx

Saved the file as a csv.

## Reading and displaying data:

```python
import pandas as pd
df = pd.read_csv("oil_consumption_data.csv", index_col = 0, na_values = "-")
colsAsInt = map(int, list(df))

df
```

## Intializing the Jupyter notebook.

```python
from bokeh.io import output_notebook

output_notebook()
```

## Generating an interactive line graph of the first 11 countries.

```python
from bokeh.plotting import figure, show
from bokeh.palettes import Spectral11

oil_data_plot = figure(plot_width=800, plot_height=400)
oil_data_plot.left[0].formatter.use_scientific = False
oil_data_plot.title.text = "Click on the legend to hide countries."

def updateGraph(countries):
    for country, color in zip(countries, Spectral11):
        y = df.loc[country]
        oil_data_plot.circle(colsAsInt, y, color = color, fill_color = "white", legend = country)
        oil_data_plot.line(colsAsInt, y, color = color, legend = country)

updateGraph(df.index)
oil_data_plot.x_range.end = 2025
oil_data_plot.legend.location = "top_right"
oil_data_plot.legend.click_policy="hide"
    
show(oil_data_plot, notebook_handle = True)
```
