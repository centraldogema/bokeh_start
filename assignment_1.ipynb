```python
import pandas as pd
df = pd.read_csv("oil_consumption_data.csv", index_col = 0, na_values = "-")
colsAsInt = map(int, list(df))

df
```

```python
import numpy as np
from ipywidgets import interact
from bokeh.io import output_notebook, push_notebook
from bokeh.layouts import widgetbox
from bokeh.models.widgets import Dropdown


output_notebook()
```

```python
from bokeh.plotting import figure, show

oil_data_plot = figure(plot_width=800, plot_height=400)
oil_data_plot.left[0].formatter.use_scientific = False

for country in df.index:
    y = df.loc[country]
    oil_data_plot.line(colsAsInt, y)
    oil_data_plot.circle(colsAsInt, y, fill_color = "white")
    
def highlight(country):
    y = df.loc[country]
    oil_data_plot.line(colsAsInt, y, color = "red")
    oil_data_plot.circle(colsAsInt, y, fill_color = "white", color = "red")
    
show(oil_data_plot, notebook_handle = True)
```

```python
interact(highlight, country=df.index.tolist())
```
